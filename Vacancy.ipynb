```python
# SOURCE: StreetEasy Rental Inventory (all)
import pandas as pd
df = pd.read_csv("rentalInventory_All/rentalInventory_All.csv")
```

```python
df = df[df["areaType"] == "neighborhood"]
```

```python
df
```

```python
# Credit to GitHub Copilot
# Extract year from column names and calculate yearly averages
df_yearly = df.filter(regex=r'^\d{4}-\d{2}$')  # Select only columns of the form "yyyy-mm"
df_yearly = df_yearly.groupby(df_yearly.columns.str[:4], axis=1).mean()

# Add back non-date columns
non_date_columns = df.drop(columns=df.filter(regex=r'^\d{4}-\d{2}$').columns)
df = pd.concat([non_date_columns, df_yearly], axis=1)
```

```python
df.drop(columns=["areaType"], inplace=True)
```

```python
df.to_csv("Vacancies.csv", index=False)
```
