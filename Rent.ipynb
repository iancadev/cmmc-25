```python
# SOURCE: StreetEasy median asking rent
# Credit to GitHub Copilot

import pandas as pd

one = pd.read_csv('medianAskingRent_OneBd/medianAskingRent_OneBd.csv')
two = pd.read_csv('medianAskingRent_TwoBd/medianAskingRent_TwoBd.csv')
three_plus = pd.read_csv('medianAskingRent_ThreePlusBd/medianAskingRent_ThreePlusBd.csv')
studio = pd.read_csv('medianAskingRent_Studio/medianAskingRent_Studio.csv')
```

```python
# Define a function to process the dataframes
def process_dataframe(df):
    # Filter rows where areaType is 'neighborhood'
    df = df[df['areaType'] == 'neighborhood']
    
    # Drop the areaType column
    df = df.drop(columns=['areaType'])
    
    # Group columns by year and calculate the average
    year_columns = [col for col in df.columns if col[:4].isdigit()]
    yearly_avg = df[year_columns].groupby(lambda x: x[:4], axis=1).mean()
    
    # Drop the original 'yyyy-mm' columns and add the yearly averages
    df = df.drop(columns=year_columns).join(yearly_avg)
    
    return df

# Process all four dataframes
one = process_dataframe(one)
two = process_dataframe(two)
three_plus = process_dataframe(three_plus)
studio = process_dataframe(studio)
```

```python
# Add a "bedrooms" column to each dataframe
studio['bedrooms'] = '0'
one['bedrooms'] = '1'
two['bedrooms'] = '2'
three_plus['bedrooms'] = '3+'

# Combine all dataframes into one
combined_df = pd.concat([studio, one, two, three_plus], ignore_index=True)
```

```python
combined_df.to_csv('medianAskingRent.csv', index=False)
```
