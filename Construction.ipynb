```python
# Source: NYC OpenData Housing Database by 2020 NTA
import pandas as pd
df = pd.read_csv("Housing_Database_by_2020_NTA.csv")
```

```python
boroughs = {
    "BK": "Brooklyn",
    "BX": "Bronx",
    "MN": "Manhattan",
    "QN": "Queens",
    "SI": "Staten Island"
}
df['borough'] = df['nta2020'].str[:2].map(boroughs)
df = df.dropna(subset=['borough'])
```

```python
df = df.drop(columns=['nta2020', 'the_geom']).rename(columns={'ntaname2020': 'areaName'})
```

```python
df
```

```python
# Rename columns with "compYYYYap" to just the year
df = df.rename(columns=lambda col: col.replace("comp", "") if "comp" in col and "ap" not in col else col)

# Keep only the columns 'borough', 'areaName', and the year columns
columns_to_keep = ['borough', 'areaName'] + [col for col in df.columns if col.isdigit()]
df = df[columns_to_keep]
```

```python
df.to_csv("Construction.csv", index=False)
```
